## Function to convert height from feet and inches to centimeters

```python
# Function to convert height from feet and inches to centimeters
def height_to_cm(height):
    # Split the height string by the apostrophe
    feet, inches = height.split("'")
    
    # Remove the double quote and convert inches to integer
    inches = inches.replace("\"", "")
    
    # Convert to total inches and then to cm
    total_inches = int(feet) * 12 + int(inches)
    return total_inches * 2.54

# Apply the function to the 'Height' column, convert the result to int, and create a new 'Height_cm' column
fifa['Height_cm'] = fifa['Height'].apply(height_to_cm).round().astype(int)
```

## Function to convert values

```python
def convert_value(value):
    # Remove the euro symbol and treat M as (Million) and K as (Thousands)
    value = value.replace('€', '').strip()  # Use 'value' instead of 'Value'
    
    # Check for M (Millions) or K (Thousands)
    if 'M' in value:
        return float(value.replace('M', '')) * 1_000_000  # Convert millions to integers
    elif 'K' in value:
        return float(value.replace('K', '')) * 1_000  # Convert thousands to integers
    else:
        return float(value)  # In case there is no M or K, just convert directly to float

# Apply the cleaning
fifa['Value'] = fifa['Value'].apply(convert_value)
```

## Function to convert wage 

```python

def convert_wage(wage):
    # Remove the euro symbol and treat M as (Million) and K as (Thousands)
    wage = wage.replace('€', '').strip()  # Use 'wage' consistently
    
    # Check for M (Millions) or K (Thousands)
    if 'M' in wage:
        return float(wage.replace('M', '')) * 1_000_000  # Convert millions to integers
    elif 'K' in wage:
        return float(wage.replace('K', '')) * 1_000  # Convert thousands to integers
    else:
        return float(wage)  # In case there is no M or K, just convert directly to float

# Apply the cleaning to the 'Wage' column
fifa['Wage'] = fifa['Wage'].apply(convert_wage)
```

```python
def convert_release_clause(release_clause):
    # Remove the euro symbol and treat M as (Million) and K as (Thousands)
    release_clause = release_clause.replace('€', '').strip()  # Remove '€' symbol and strip any leading/trailing spaces
    
    # Check for M (Millions) or K (Thousands)
    if 'M' in release_clause:
        return float(release_clause.replace('M', '')) * 1_000_000  # Convert millions to integers
    elif 'K' in release_clause:
        return float(release_clause.replace('K', '')) * 1_000  # Convert thousands to integers
    else:
        return float(release_clause)  # In case there is no M or K, just convert directly to float

# Apply the cleaning function to the 'Release Clause' column
fifa['Release Clause'] = fifa['Release Clause'].apply(convert_release_clause)
```
